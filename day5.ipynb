```python
f=open('day5_input1.txt')
lines=f.readlines()

seeds=lines[0].split(':')[1]
seeds=seeds.split()
print('seeds',seeds)

d={}
for line in lines[2:]:
    if line.find('map')!=-1:
        map_name = line[:line.find('map')].strip()
        d.setdefault(map_name,[])
    else:
        row_list = line.split()
        if row_list:
            d.setdefault(map_name,[]).append(row_list)
```

```python
trans_list=[]
for seed in seeds:
    # print('----------------------------')
    # print(seed,'to be transformed')
    
    # seed-to-soil transformation
    initial=int(seed)
    transformed=0
    for i in d['seed-to-soil']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Soil:',initial)

    # soil-to-fertilizer transformation
    for i in d['soil-to-fertilizer']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Fertilizer:',initial)

    # fertilizer-to-water transformation
    for i in d['fertilizer-to-water']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Water:',initial)

    # water-to-light transformation
    for i in d['water-to-light']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Light:',initial)

    # light-to-temperature transformation
    for i in d['light-to-temperature']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Temp:',initial)

    # temperature-to-humidity transformation
    for i in d['temperature-to-humidity']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Humidity:',initial)

    # humidity-to-location transformation
    for i in d['humidity-to-location']:
        if (initial>=int(i[1]))&(initial<=(int(i[1])+int(i[2])-1)):
            diff = abs(initial - int(i[1]))
            transformed = int(i[0])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Location:',initial)
    trans_list.append(initial)

print(min(trans_list))
```

```python
from tqdm import tqdm

f=open('day5_input1.txt')
lines=f.readlines()

seeds=lines[0].split(':')[1]
seeds=seeds.split()
n=2
part_seed = [seeds[i:i+n] for i in range(0,len(seeds),n)]

d={}
for line in lines[2:]:
    if line.find('map')!=-1:
        map_name = line[:line.find('map')].strip()
        d.setdefault(map_name,[])
    else:
        row_list = line.split()
        if row_list:
            d.setdefault(map_name,[]).append(row_list)
```

```python
for loc in tqdm(range(10000000,100000001)):
    # print('----------------------------')
    # print('Location',loc,'to be transformed')

    # humidity-to-location transformation
    initial=int(loc)
    transformed=0
    for i in d['humidity-to-location']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Humidity:',initial)

    # temperature-to-humidity transformation
    for i in d['temperature-to-humidity']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Temperature:',initial)
    
    # light-to-temperature transformation
    for i in d['light-to-temperature']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Light:',initial)

    # water-to-light transformation
    for i in d['water-to-light']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Water:',initial)

    # fertilizer-to-water transformation
    for i in d['fertilizer-to-water']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Fertilizer:',initial)

    # soil-to-fertilizer transformation
    for i in d['soil-to-fertilizer']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Soil:',initial)

    # seed-to-soil transformation
    for i in d['seed-to-soil']:
        if (initial>=int(i[0]))&(initial<=(int(i[0])+int(i[2])-1)):
            diff = abs(initial - int(i[0]))
            transformed = int(i[1])+diff
    if transformed!=0:
        initial=transformed
        transformed=0
    # print('Seed:',initial)

    for ps in part_seed:
        if (initial>=int(ps[0]))&(initial<=(int(ps[0])+int(ps[1])-1)):
            print(loc)
            break
    else:
        continue
    break
```
